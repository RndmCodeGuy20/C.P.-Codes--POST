```python
import matplotlib.pyplot as pl
import numpy as np
x = np.arange(0,10,0.2)
a=np.cos(x)
b=np.sin(x)
pl.plot(x,a,'r')
pl.plot(x,b,'b')
pl.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6,3,5,2,1,1,6,3,5]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86,24,36,48,18,45,23,67,2]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
def factorial(k):
    if (k>0):
        result = k * factorial(k-1)
        print(result)
    else:
        result=1
    return result

print("Factorial of the input number is")
factorial(3)
```
